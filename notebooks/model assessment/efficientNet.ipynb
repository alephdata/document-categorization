```python
import tensorflow as tf
from tensorflow import keras
```

```python
modelB0 = tf.keras.applications.efficientnet.EfficientNetB0(include_top=True, weights=None, classes=4)
print(modelB0.summary())
```

```python
modelB1 = tf.keras.applications.efficientnet.EfficientNetB1(include_top=True, weights=None, classes=4)
print(modelB1.summary())
```

```python
modelB2 = tf.keras.applications.efficientnet.EfficientNetB2(include_top=True, weights=None, classes=4)
print(modelB2.summary())
```

```python
modelVGG = tf.keras.applications.vgg16.VGG16(include_top=True, weights=None, classes=4)
print(modelVGG.summary())
```

```python
modelAN = keras.models.Sequential(
    [
        keras.layers.Conv2D(filters=96, kernel_size=(11, 11), strides=(4, 4), activation="relu", input_shape=(227, 227, 3)),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
        keras.layers.Conv2D(filters=256, kernel_size=(5, 5), strides=(1, 1), activation="relu", padding="same"),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
        keras.layers.Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
        keras.layers.BatchNormalization(),
        keras.layers.Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
        keras.layers.BatchNormalization(),
        keras.layers.Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation="relu", padding="same"),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
        keras.layers.Flatten(),
        keras.layers.Dense(4, activation="softmax"),
    ]
)
print(modelAN.summary())
```
